```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Sample movie ratings data
data = {
    'Movie': ['Movie A', 'Movie B', 'Movie C', 'Movie D', 'Movie E', 'Movie F', 'Movie G', 'Movie H'],
    'Genre': ['Action', 'Comedy', 'Drama', 'Action', 'Drama', 'Comedy', 'Action', 'Drama'],
    'Rating': [4.5, 3.0, 4.0, 5.0, 2.5, 3.5, 4.8, 3.8],
    'Votes': [1500, 800, 1200, 2000, 400, 700, 1800, 1100]
}

df = pd.DataFrame(data)

# Show basic info
print(df.head())

# Ratings distribution
plt.figure(figsize=(6,4))
sns.histplot(df['Rating'], bins=5, kde=True)
plt.title('Movie Ratings Distribution')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.show()

# Top 3 movies by rating
top_movies = df.sort_values(by='Rating', ascending=False).head(3)
print("\nTop 3 Movies by Rating:")
print(top_movies[['Movie', 'Rating']])

# Barplot of top movies
plt.figure(figsize=(8,4))
sns.barplot(data=top_movies, x='Movie', y='Rating', palette='viridis')
plt.title('Top 3 Movies by Rating')
plt.ylim(0, 5)
plt.show()
```
